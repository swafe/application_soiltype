```
#%% change rotedir and path_dst_folder
f = open('/content/data/vgg_annotation/img_horizontal/val_5206_5272/recognice_sample_val_5206_5272_edit_edit.json',)
root_dir_img_folder = '/content/data/vgg_annotation/img_horizontal/val_5206_5272'
path_json_edit = '/content/data/vgg_annotation/img_horizontal/val_5206_5272/recognice_sample_val_5206_5272_edit_edit2.json'
# f = open('/content/data/vgg_annotation/img_horizontal/train_0247_5205/recognice_sample_train_5205_edit_edit.json',)
# root_dir_img_folder = '/content/data/vgg_annotation/img_horizontal/train_0247_5205'
# path_json_edit = '/content/data/vgg_annotation/img_horizontal/train_0247_5205/recognice_sample_train_5205_edit_edit2.json'


data = json.load(f)
```

```
data_keys_json = list(data.keys())#['IMG_0332.JPG1076716']
# data_basename_img_file
for img_key in data_keys_json:
  # print(img_key)
  # print(data[img_key]['size'])
  data_filename = data[img_key]['filename']
  # print(data_filename)

  image_path = os.path.join(root_dir_img_folder, data_filename)
  img_size_horiz =  os.path.getsize(image_path)
  filename_and_size_horiz = data_filename + str(img_size_horiz)
  # print(filename_and_size_horiz)
  data[img_key]['size'] = img_size_horiz
  # print(data[img_key]['size'])
  data[img_key]['size'] = img_size_horiz

  data[filename_and_size_horiz]=data.pop(img_key)
  #print(list(data.keys())[0])
with open(path_json_edit, 'w') as fp:
        json.dump(data, fp)
```

```
#alle img
data_keys_json = list(data.keys())
sample = {"sample": "sample"}
for i_data_keys_json in data_keys_json:
  #alle regions
  print(i_data_keys_json)
  for i_shape_attributes in data[i_data_keys_json]['regions']:
    print(i_shape_attributes['region_attributes'])
    if not i_shape_attributes['region_attributes']:
      i_shape_attributes['region_attributes'].update(sample)
      test =i_shape_attributes['region_attributes']['sample']
      print('created' +test)
    else:
      print(i_shape_attributes['region_attributes']['sample'])
with open(path_json_edit, 'w') as fp:
        json.dump(data, fp)
```
